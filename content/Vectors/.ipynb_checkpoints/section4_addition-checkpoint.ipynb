## Adding a point and a vector
---

Consider a point $A(A_x \vert A_y \vert A_z)$ and a vector $$\vec v =\left(\begin{array}{r} v_x\\v_y\\v_z \end{array}\right)$$

If we start at point $A$ and follow the instructions given by vector $\vec v$, we end up at new point. We denote this point by

$$\boxed{A+\vec v}$$

The coordinates of this new point are simply given by summing the coordinates of $A$ and the components of $v$:

$$ \boxed{(A_x+v_x \vert A_y+v_y \vert A_z + v_z)} $$

Sometimes we give this new point a name, for example 

$$B = A+\vec v$$
